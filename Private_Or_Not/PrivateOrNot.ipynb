```python
dataschool = pd.read_csv('../Grades/dataSet.csv')
dataschool
```

```python
def isPrivate(institute,localidad):
    p = bac[(bac.NOM_CENTRO == institute) & (bac.LOCALIDAD == localidad)].head(1).REG_JUR.to_json()
    if(p == '{}'):
        p = bac[(bac.NOM_CENTRO == institute)].head(1).REG_JUR.to_json()
    
    pos = p.find('P')
    p = p[pos:pos+3]
    if(p == 'PRI'):
        return True 
    elif(p == 'PUB'):
        return False
    elif(institute.find('PRIVAT')!=-1 or institute.find('PRIVADO')!=-1):
        return True
    else: 
        return 'Null'

# I set True if the school is private, false if not 
private = []
for school in dataschool.itertuples():
    private.append(isPrivate(school.Instituto,school.Localidad))

dataschool['private']=private
dataschool
```

```python
# I assume that the remaining ones are Public because are more common 
dataschool['private'] = dataschool['private'].replace(['Null'],False)
dataschool
```
