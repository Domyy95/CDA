```python
import folium
import json
```

```python
with open('map.geojson') as f:
    data = json.load(f)

m = folium.Map(location=[39.4697500, -0.3773900])
m
```

```python
folium.features.GeoJson(
    'map.geojson'
).add_to(m)

for each in data['features']:
    if(each['geometry']['type'] == 'Point'):
        folium.Marker([each['geometry']['coordinates'][1],each['geometry']['coordinates'][0]], popup=each['properties']['Name']).add_to(m)
        
```

```python
m.save('map.html')
m
```
