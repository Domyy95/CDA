```python
import pandas as pd
import numpy as np

data = pd.read_csv('income_per_section/dataSet_3.csv',index_col=0)
dataprivate = pd.read_csv('Private_Or_Not/dataSet_2.csv')
                                                
# Data Cleaning
data['private'] = dataprivate.private
data
```

```python
data = data.dropna()
data = data.drop(columns = 'coordinate')
data = data.drop(columns = 'Presentados')
data = data.drop(columns = 'Aptos')
data = data.drop(columns = 'lat')
data = data.drop(columns = 'long')
data
```

```python
data[['Media Expediente','Media PAU','NumBib 2 km','NumBib 5 km','NumBib 8 km','NumBib 10 km','private']].corr(method="kendall")
```

```python
from sklearn import linear_model
import sklearn
msk = np.random.rand(len(data)) < 0.8
msk
train = data[msk]
test = data[~msk]

X_train = train[['private','income','NumBib 2 km']]
Y_train = train['Media Expediente']

Y_test = test[['private','income','NumBib 2 km']]

regr = linear_model.LinearRegression()
regr.fit(X_train, Y_train)
prediction = regr.predict(test[['private','income','NumBib 2 km']])
print('Coefficients: \n', regr.coef_)
# The mean squared error
print('Mean Abs Error: ' + str(sklearn.metrics.mean_absolute_error(test[['Media Expediente']], prediction)))
```

```python
import matplotlib.pyplot as plt
plt.plot(test[['Media Expediente']],prediction,'bo')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(X_train,Y_train)
regr_2.fit(X_train,Y_train)

y1 = regr_1.predict(Y_test)
y2 = regr_2.predict(Y_test)
```

```python
plt.plot(test[['Media Expediente']], y1, 'bo',color="blue",
         label="max_depth=2", linewidth=2)
plt.plot(test[['Media Expediente']], y2,'bo', color="green", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

```python
print('Mean Abs Error: ' + str(sklearn.metrics.mean_absolute_error(test[['Media Expediente']], y1)))
```
